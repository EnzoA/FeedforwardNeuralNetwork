```python
import numpy as np
from feedforward_neural_network import FeedforwardNeuralNetwork, OptimizationAlgorithm
from matplotlib import pyplot as plt
```

```python
neural_network = FeedforwardNeuralNetwork(architecture=(2, 2, 1))
    
losses = neural_network.fit(
    X=np.array([[0, 0], [0, 1], [1, 0], [1, 1]]),
    y=np.array([0, 1, 1, 0]),
    epochs=20000,
    lr=0.1,
    optimization_algorithm=OptimizationAlgorithm.STOCHASTIC_GRADIENT_DESCENT,
    verbose=True
)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(losses)
plt.show()
```

```python
print(
    f'XOR prediction for 0, 0 is {int(np.rint(neural_network.predict(np.array([0, 0]))[0][0]))}'
)

print(
    f'XOR prediction for 0, 1 is {int(np.rint(neural_network.predict(np.array([0, 1]))[0][0]))}'
)

print(
    f'XOR prediction for 1, 0 is {int(np.rint(neural_network.predict(np.array([1, 0]))[0][0]))}'
)

print(
    f'XOR prediction for 1, 1 is {int(np.rint(neural_network.predict(np.array([1, 1]))[0][0]))}'
)
```
